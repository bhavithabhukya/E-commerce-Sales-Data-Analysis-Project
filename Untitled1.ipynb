```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default="plotly_white"
```

```python
df=pd.read_csv("superstore.csv")
print("Initial Shape:", df.shape)
print(df.head())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Fill missing categories with "Miscellaneous"
if 'Category' in df.columns:
    df['Category'] = df['Category'].fillna('Miscellaneous')

# Convert 'Order Date' or 'Date' to datetime
if 'Order Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Order Date'])
elif 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Create Amount column (use Sales if Amount not present)
if 'Sales' in df.columns:
    df['Amount'] = df['Sales']

# Drop rows missing critical values
df.dropna(subset=['Date', 'Amount'], inplace=True)

print("\nCleaned Data Shape:", df.shape)
```

```python
if 'City' in df.columns:
    city_sales = df.groupby('City')['Amount'].agg(['sum', 'mean']).reset_index()
    city_sales.columns = ['City', 'Total_Sales', 'Average_Order_Value']
    print("\nQ1: Total and Average Sales by City:\n")
    print(city_sales.sort_values(by='Total_Sales', ascending=False).head(10))
else:
    print("No 'City' column found.")

```

```python
if 'Category' in df.columns:
    category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    print("\nQ2: Best Selling Product Categories:\n")
    print(category_sales)
else:
    print("No 'Category' column found.")
```

```python
# Already handled in cleaning step
df['Category'] = df['Category'].fillna('Miscellaneous')
print("✅ Missing categories replaced with 'Miscellaneous'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extract month and quarter
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = df['Date'].dt.to_period('Q')

# Monthly and Quarterly Sales
monthly_sales = df.groupby('Month')['Amount'].sum().sort_values(ascending=False)
quarterly_sales = df.groupby('Quarter')['Amount'].sum()

print("\nQ4: Monthly Sales:\n", monthly_sales)
print("\nQ4: Quarterly Sales:\n", quarterly_sales)

# Plot Monthly Trend
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
if 'Customer ID' in df.columns:
    top_customers = df.groupby('Customer ID')['Amount'].sum().sort_values(ascending=False).head(10)

    plt.figure(figsize=(10,5))
    sns.barplot(x=top_customers.index, y=top_customers.values, palette='coolwarm')
    plt.title("Top 10 Customers by Total Purchase")
    plt.xlabel("Customer ID")
    plt.ylabel("Total Purchase Amount")
    plt.xticks(rotation=45)
    plt.show()
```

```python
if 'City' in df.columns:
    heatmap_data = df.pivot_table(values='Amount', index='City', columns='Month', aggfunc='sum', fill_value=0)

    plt.figure(figsize=(12,6))
    sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".0f")
    plt.title("Sales Heatmap by City and Month")
    plt.show()
```

```python
print("""
✅ Summary of Findings:
- Total and average sales calculated for each city
- Best-selling product categories identified
- Missing categories handled as 'Miscellaneous'
- Monthly and quarterly sales trends visualized
- Top 10 customers and city-month heatmap created
""")
```
